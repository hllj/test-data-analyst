# Các phép tính trong Python

Trong python tồn tại các phép tính để cộng, trừ, nhân, chia, ...

Tuy nhiên có một số chú ý, đầu tiên là cộng trừ nhân chia với các số nguyên hoặc số thực.

Cộng (+), Trừ (-), Nhân (*), Chia (/)

```python
print(2 + 3)
print(1 - 10)
print(1 * 5)
print(5 / 2)
```

```python
a = 2
b = 3
a * b
```

Tồn tại một dạng khác là phép chia lấy phần nguyên sẽ dùng dấu "//". Cái này cần chú ý.

```python
print("5 / 2:", 5 / 2)
print("5 // 2:", 5 // 2)
```

Các phép tính cũng có thể áp dụng cho các chuỗi, ví dụ:
- Chuỗi chỉ cho phép cộng 2 chuỗi -> Đây là phép nối 2 chuỗi lại với nhau.
- Không có thể trừ, nhân, chia 2 chuỗi

```python
"123" + "456"
```

```python
"123" - "456"
```

Tương tự, ta có thể dùng phép cộng để nối 2 list, 2 tuple.

```python
[1, 2, 3] + ["a", "b", "c"]
```

```python
(10, 2.5, "abc") + (1, 2)
```

```python
{"a": 1} + {2: "b"}
```
