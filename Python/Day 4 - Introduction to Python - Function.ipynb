```python
# Function with parameters and return value
def add_numbers(a, b):
  return a + b

result = add_numbers(5, 3)
print(result)  # Output: 8
```

```python
# Function to find the square of a number
def square(num):
    return num * num

print(square(4))  # Output: 16
```

```python
print(square(3))
```

```python
# Function to print each item in a list
def print_items(items):
    for item in items:
        print(item)

print_items(["apple", "banana", "cherry"])
```

```python
# Function to count down from a number
def countdown(start):
    while start > 0:
        print(start)
        start -= 1
    print("Liftoff!")

countdown(5)
```

```python
# Function with default parameter value
def greet_user(name="Guest"):
    print(f"Hello, {name}!")

greet_user("Alice")  # Output: Hello, Alice!
greet_user()          # Output: Hello, Guest!
```

```python
# Function to filter even numbers from a list
def filter_evens(numbers):
    evens = []
    for num in numbers:
        if num % 2 == 0:
            evens.append(num)
    return evens

print(filter_evens([1, 2, 3, 4, 5, 6]))  # Output: [2, 4, 6]
```
