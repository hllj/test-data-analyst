# For loop:
For loop được sử dụng khi ta lặp các phần tử trong một cấu trúc dữ liệu nào đó list, tuple, dictionary, range, ...

Ta biết được số lần lặp.

Dưới đây là một số ví dụ sử dụng vòng lặp for trong các tình huống khác nhau:

For loop trong 1 list

```python
fruits = ["apple", "banana", "cherry"] # Lặp trong một list
for fruit in fruits:
    print(fruit)
```

For loop trong một khoảng với range()

```python
list(range(10))
```

```python
list(range(1, 10))
```

```python
for i in range(5): # từ 0 tới 4
  print(i)
```

```python
for i in range(1, 10): # từ 1 tới 9
  print(i)
  if i % 2 == 0:
    print("Đây là số chẵn")
  else:
    print("Đây là số lẻ")
```

```python
for i in range(10, 101, 10): # đi từ 10 tới 100, cách nhau 10 đơn vị, nó sẽ 1 mảng từ 10 -> 100, mỗi giá trị cách nhau 10 đơn vị
  print(i)
```

Lặp với các phần tử trong dictionary với

```python
person = {"name": "Alice", "age": 25, "city": "New York"}
```

```python
person.items()
```

```python
for k, v in person.items():
  print("Key:", k, "Value:", v)
```

```python
for k in person:
  print("Key:", k)
```

```python
for k in person:
  print("Key:", k, "with value is:", person[k])
```

Ta cũng có thể làm ý tưởng tương tự khi lặp các số trong một list bằng cách lặp các chỉ số:

ví dụ:
- list_a = ["a", "b", "c"], lặp từ 0, 1, 2, sử dụng hàm len() và range().

```python
l = ["a", "b", "c"]

for i in range(len(l)): # i = 0, 1, 2 vì len(list_a) == 3 nên i sẽ lấy trong khoảng range(3)
  print("Phần tử thứ: ", i, " có giá trị: ", l[i])
```

```python
len(l)
```

```python
range(len(l))
```

Dừng vòng lặp với break

```python
for i in range(10): #Từ 0 tới 9
    if i == 5: # nếu gặp i == 5 thì thoát ra luôn khỏi vòng lặp bằng break
        break
    print(i)
```

```python
for i in range(10): #Từ 0 tới 9
    if i == 5:
        continue
    print(i)
```

```python
# i = 0 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 0
# i = 1 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 1
# i = 2 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 2
# i = 3 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 3
# i = 4 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 4
# i = 5 -> i == 5 -> đúng -> continue ->
# i = 6 -> i == 5 -> sai (không đi qua continue) -> print(i) -> 6
```

Tiếp tục vòng lặp với continue:

Ví dụ:
Lặp các số từ 1 tới 10, nếu gặp một số chẵn thì bỏ qua

```python
for i in range(1, 11):
  if i % 2 == 0:
    continue
  print("số lẻ: ", i)
```

# While loop

```python
sum = 0
while sum <= 100:
  x = int(input("Nhập một số:"))
  sum += x
  print("Tổng hiện tại là: ", sum)

print("Tổng đã nhập là:", sum)
```

Lặp các số từ 1 và bé hơn hoặc bằng 5 mỗi lần cộng thêm 2 đơn vị.

```python
i = 1
while i <= 5:
    print(i)
    i += 2
```

```python
print(i)
```

Khởi tạo i = 1

Vô vòng lặp:

- i == 1 <= 5 -> Đúng -> print(i) 1 -> in ra 1 -> i += 2 -> i = 3
- i == 3 <= 5 -> Đúng -> print(i) 3 -> in ra 3 -> i += 2 -> i = 5
- i == 5 <= 5 -> Đúng -> print(i) 5 -> in ra 5 -> i += 2 -> i = 7
- i == 7 <= 5 -> Sai -> Thoát khỏi vòng lặp

i cuối cùng 7

Nhập các số tự nhiên cho tới khi tổng các số lớn hơn 100 thì dừng lại.

```python
total = 0

while True:
    i = int(input("Nhập một số:"))
    total = total + i
    print("Tổng hiện tại là: ", total)
    if total > 100:
      break
print("Tổng các số là:", total)
```

Xoá các phần tử trong mảng:

```python
tasks = ["Học Python", "Làm bài tập", "Ôn lại kiến thức"]
while len(tasks) > 0:
    print("Đang làm:", tasks.pop(0))
    print("Các việc còn lại:", tasks)

print("Tasks: ", tasks)
```

Nhập liên tục các số, thêm nó vào trong một mảng cho tới khi số lượng mảng của nó là 5.

```python
numbers = []

while len(numbers) < 5:
  x = int(input("Nhập vào một số:"))
  numbers.append(x)

print("Mảng vừa nhập:", numbers)
```

```python
numbers = []

while True:
  x = int(input("Nhập vào một số:"))
  numbers.append(x)
  if len(numbers) == 5:
    break

print("Mảng vừa nhập:", numbers)
```
