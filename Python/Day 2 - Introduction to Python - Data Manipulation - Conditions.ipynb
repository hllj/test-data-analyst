## Dictionary

## Create a Dictionary

```python
# create dictionary using { }
d1 = {1: 'Geeks', 2: 'For', 3: 'Geeks'}
print(d1)

# create dictionary using dict() constructor
d2 = dict(a = "Geeks", b = "for", c = "Geeks")
print(d2)
```

## Read a Dictionary

Có 2 cách để đọc một key ở trong Dictionary:
- dict[tên key] -> trả về value tương ứng.
- dict.get(tên key).

```python
d = { "name": "Alice", 1: "Python", (1, 2): [1,2,4] }

# Access using key
print(d["name"])

# Access using get()
print(d.get("name"))
```

```python
d["age"]
```

## Update:
- Cách để thêm một key-value vào dictionary. dict[key] = value
- Cách để thay đổi giá trị key-value trong dictionary thành giá trị value mới. dict[key] = value mới.

```python
d = {1: 'Geeks', 2: 'For', 3: 'Geeks'}

# Adding a new key-value pair
d["age"] = 22
```

```python
d
```

```python
# Updating an existing value
d[1] = "Python dict"

print(d)
```

## Delete

```python
d = {1: 'Geeks', 2: 'For', 3: 'Geeks', 'age':22}

# Using del to remove an item
del d["age"] # xoá bằng del
print(d)
```

```python
# Using pop() to remove an item and return the value
val = d.pop(1) # đưa giá trị key cần xoá
print(val)
```

```python
d
```

```python
# Using popitem to removes and returns
# the last key-value pair.
key, val = d.popitem()
print(f"Key: {key}, Value: {val}")
```

```python
d
```

```python
# Clear all items from the dictionary
d.clear() # Xoá toàn bộ các key-value ở trong dictionary
print(d)
```

## So sánh với nhiều luồng xử lý

Để xử lý điều kiện có nhiều trường hợp (3 trở lên) thì ta sẽ sử dụng if - elif - else:

if <điều kiện số 1>:

  <xử lý cho điều kiện 1 đúng>

elif <điều kiện số 2>:

  <xử lý cho điều kiện 2 đúng>

elif <điều kiện số 3>:

  <xử lý cho điều kiện 3 đúng>

elif <điều kiện thứ n>:

  <xử lý cho điều kiện n đúng>

else:

  <xử lý cho điều kiện không thoả mãn tất cả từ 1 tới n>

```python
month = 1

season = None

if month in [1, 2, 3]:
  print("Mùa xuân")
  season = "Spring"
elif month in [4, 5, 6]:
  print("Mùa hè")
  season = "Summer"
elif month in [7, 8, 9]:
  print("Mùa thu")
  season = "Fall"
elif month in [10, 11, 12]:
  print("Mùa đông")
  season = "Winter"
else:
  print("Tháng không hợp lệ")

season
```

```python
month = 1

if 1 <= month <= 3:
  print("Mùa xuân")
elif 4 <= month <= 6:
  print("Mùa hè")
elif 7 <= month <= 9:
  print("Mùa thu")
elif 10 <= month <= 12:
  print("Mùa đông")
else:
  print("Tháng không hợp lệ")
```

So sánh bằng: a == b -> toàn bộ các cấu trúc dữ liệu như số nguyên, số thực, chuỗi, list, ...

So sánh lớn hơn: >
So sánh lớn hơn hoặc bằng: >=

So sánh nhỏ hơn: <
So sánh bé hơn hoặc bằng: <=

```python
a = "abc"
b = "abc"
a == b
```

```python
a = 10
b = 20
a < b
```

Ta có thể tạo ra các điều kiện ghép nhau:

- and: và; <điều kiện 1> and <điều kiện 2>,
- or: hoặc; <điều kiện 1> or <điều kiện 2> or ... <điều kiện thứ n>

```python
a = 10
b = 20

if (a <= b) and (a > 0):
  print("Số a nhỏ hơn số b và số a lớn hơn 0")
```

```python
city = "HCMC"

region = "Europe"

population = 1

if (region == "Asia") or (population >= 10) or (city == "HCMC"):
  print("Thành phố này ở châu Á hoặc dân số lớn hơn 10 triệu người hoặc là thành phố Hồ Chí Minh")
```
